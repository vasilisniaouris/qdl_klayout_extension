```python
# import packages
import warnings

from qdl_klayout_extension.core.shapes import RingPath, ArcPath, SimpleCircle
from qdl_klayout_extension.core.layout import Layout
from qdl_klayout_extension.visualize import plot_shape

warnings.filterwarnings("ignore")
```

```python
layout = Layout()
layer = layout.layer(1, 0)
```

```python
base_cell = layout.create_cell('base')
```

```python
def get_shape_list(circle_radius, ring_radius, ring_width, arc_radius, arc_width):
    circle = SimpleCircle.from_center(circle_radius)
    ring = RingPath.from_center(ring_radius, ring_width)
    arc1 = ArcPath.from_center(arc_radius, arc_width, start_angle=45, angle_range=60)
    arc2 = ArcPath.from_center(arc_radius, arc_width, start_angle=125, angle_range=60)
    arc3 = ArcPath.from_center(arc_radius, arc_width, start_angle=205, angle_range=60)

    return [circle, ring, arc1, arc2, arc3]


shape_list = get_shape_list(5, 8, 1, 10, 1)
for shape in shape_list:
    plot_shape(shape)
```

```python
base_cell.easy_insert(layer, shape_list)
```

```python
layout.write('example_4_concentric_shapes.gds')
```
